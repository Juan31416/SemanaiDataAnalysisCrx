```python
df = pd.read_csv('sample_semanai.csv')
```

```python
df = df.drop(columns = ['key'])
```

```python
gByOrder = df.groupby(["orden", "subGroup"]).size().reset_index(name='count')
basket = (gByOrder.groupby(["orden", "subGroup"])['count']
          .sum().unstack().reset_index().fillna(0)
          .set_index('orden'))
```

```python
#The encoding function
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
```

```python
custom_basket = basket_sets[['BEBIDAS CON ALCOHOL', 'HAMBURGUESAS','BÁSICOS','CLÁSICOS']]
```

```python
basket_sets.columns
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift")
rules.sort_values('confidence', ascending = False, inplace = True)
rules.head(10)
```

```python
frequent_itemsets = apriori(custom_basket, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift")
rules.sort_values('confidence', ascending = False, inplace = True)
```
